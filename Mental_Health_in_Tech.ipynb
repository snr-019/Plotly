#Data Visualization Using Plotly

```python
#importing libraries
import plotly.express as px
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objs as go
import plotly.io as pio
pio.renderers
```

```python
#loading dataset
df=pd.read_csv("/content/drive/MyDrive/Data Sets/survey.csv")
df
```

```python
print(df['Country'].value_counts())
print("\n \n")
print(df['state'].unique())
```

```python
# Dropping
#print(df.columns)
df = df.drop(columns=['state', 'comments', 'Timestamp'])
print('Columns in updated Dataframe :',len(df.columns))
```

```python
print("The dataset contains different age groups including: \n")
print(df['Age'].unique())
print("\n \n")
print("The different gender notations used in our dataset are: \n")
print(df['Gender'].unique())
```

```python
df.drop(df[df['Age'] < 0].index, inplace = True)
df.drop(df[df['Age'] > 100].index, inplace = True)
df['Age'].unique()
```

```python
df['Gender'].replace(['Male ', 'male', 'M', 'm', 'Male', 'Cis Male',
                     'Man', 'cis male', 'Mail', 'Male-ish', 'Male (CIS)',
                      'Cis Man', 'msle', 'Malr', 'Mal', 'maile', 'Make',], 'Male', inplace = True)

df['Gender'].replace(['Female ', 'female', 'F', 'f', 'Woman', 'Female',
                     'femail', 'Cis Female', 'cis-female/femme', 'Femake', 'Female (cis)',
                     'woman',], 'Female', inplace = True)

df["Gender"].replace(['Female (trans)', 'queer/she/they', 'non-binary',
                     'fluid', 'queer', 'Androgyne', 'Trans-female', 'male leaning androgynous',
                      'Agender', 'A little about you', 'Nah', 'All',
                      'ostensibly male, unsure what that really means',
                      'Genderqueer', 'Enby', 'p', 'Neuter', 'something kinda male?',
                      'Guy (-ish) ^_^', 'Trans woman',], 'Other', inplace = True)

df['Gender'].value_counts()
```

```python
# Distribution of respondents by Country
fig_country = px.histogram(df, x='Country', title='Distribution of Countries and Respondents',
                           color_discrete_sequence=['darkmagenta'])
fig_country.update_layout(xaxis_title='Country', yaxis_title='Respondent Count',
                         title_font=dict(size=30, family='Arial', color='red'),
                         xaxis=dict(title=dict(font=dict(size=20, family='Arial', color='black'))),
                         yaxis=dict(title=dict(font=dict(size=20, family='Arial', color='black'))),
                         title_font_size=30, title_x=0.5,)
fig_country.update_xaxes(tickangle=45)
```

Insights:We can easily determine which countries have the highest and lowest representation in the dataset. If some countries have very low counts or are absent, this could be an indication of geographical bias in the data collection.

```python
#distribution of age
fig = px.box(df["Age"] ,title='Age Distribution in Survey')
fig.update_layout(
    xaxis_title='Age',
    yaxis_title='Frequency',
    title_x=0.5,
    template='plotly_dark',
    title_font=dict(size=30, family='Arial', color='red'),
    xaxis=dict(title=dict(font=dict(size=20, family='Arial', color='white'))),
    yaxis=dict(title=dict(font=dict(size=20, family='Arial', color='white'))),
)
fig.show()
```

Insights:It provides a snapshot of the age distribution, allowing you to identify the spread, and potential outliers.

```python
# Plot for Gender Distribution
fig_gender = px.histogram(df, x='Gender', title='Gender Distribution in Survey')
fig_gender.update_layout(xaxis_title='Gender', yaxis_title='Count', template='plotly_dark')
fig_gender.show()
```

```python
# Distribution of Age with gender
fig_age = px.histogram(df, x='Age', color='Gender', title='Age Distribution by Gender in Survey',  opacity=0.8,barmode='overlay' )
fig_age.update_layout(xaxis_title='Age', yaxis_title='Count',
                      title_font=dict(size=30, family='Arial', color='red'),
                      xaxis=dict(title=dict(font=dict(size=20, family='Arial', color='black'))),
                      yaxis=dict(title=dict(font=dict(size=20, family='Arial', color='black'))),)
```

Insights:The plot visualizes the distribution of ages among survey respondents, with a separate color for each gender. The aim was to identify any age-specific patterns in the dataset and observe if there's a significant difference in age distribution between genders

```python
# distribution of family history
fig_country = px.histogram(df, x='family_history', title='Family History of Mental Health',
                           color_discrete_sequence=['darkmagenta'])
fig_country.update_layout(xaxis_title='Family-History', yaxis_title='Count',
                         title_font=dict(size=30, family='Arial', color='aqua'),
                         xaxis=dict(title=dict(font=dict(size=20, family='Arial', color='white'))),
                         yaxis=dict(title=dict(font=dict(size=20, family='Arial', color='white'))),
                         title_font_size=30, title_x=0.5,
                          template='plotly_dark',
                         bargap=0.8)
```

Insights: Shows the family histroy whether the employee has someone from the past involved in mental illness

```python
#  pie chart for Treatment
fig_treatment = px.pie(df, names='treatment', title='Proportion of Respondents Seeking Treatment',
                       color_discrete_sequence=px.colors.sequential.RdBu)
fig_treatment.update_traces(textinfo='percent+label', pull=[0.010,0],hole=0.4)
fig_treatment.update_layout(title_font_size=20, title_x=0.5, template='plotly_white',
                           legend_title_text='Treatment Sought',
                           legend=dict( yanchor='bottom', y=0.02, xanchor='auto', traceorder="normal"))
fig_treatment.show()
```

Insights:chart visually demonstrates the proportion of respondents who have sought treatment for mental health issues compared to those who have not.

```python
df_pie = df['work_interfere'].value_counts().reset_index()
df_pie.columns = ['work_interfere', 'count']
fig_pie = px.pie(
    df_pie,
    names='work_interfere',
    values='count',
    title='Frequency of Mental Health Issues Interfering with Work',
    color_discrete_sequence=px.colors.qualitative.Dark24
)
fig_pie.update_traces(textinfo='percent+label',hole=0.4, pull=[0.03,0])
fig_pie.update_layout(
    title_font=dict(size=30, family='Arial', color='red'),
    title_x=0.5,  # Center the title
    legend_title_font=dict(size=15, family='Arial'),
    legend_font=dict(size=15, family='Arial'),
    showlegend=True,

)
fig_pie.show()
```

Insights:the chart visually highlights the extent to which mental health issues can interfere with work, providing a basis for deeper analysis and workplace policy considerations.

```python
df_grouped = df.groupby(['Gender', 'work_interfere']).size().reset_index(name='counts')
fig_gender_work_interference = px.bar(
    df_grouped,
    x='Gender',
    y='counts',
    color='work_interfere',
    title='Relationship Between Gender and Work Interference',
    barmode='group',
    color_discrete_sequence=px.colors.qualitative.G10,
)
fig_gender_work_interference.update_layout(
    title_font=dict(size=30, family='Arial', color='aqua'),
    title_x=0.5,
    template='plotly_dark',
    xaxis=dict(
        title_font=dict(size=20, family='Arial', color='white'),
        tickfont=dict(size=15, family='Arial'),
    ),
    yaxis=dict(
        title_font=dict(size=20, family='Arial', color='white'),
        tickfont=dict(size=15, family='Arial')
    ),
    showlegend=True,
    legend=dict(
        title_font=dict(size=15, family='Arial', color='white'),
        font=dict(size=15, family='Arial'),
        bordercolor='black',
        borderwidth=1,
    ))
fig_gender_work_interference.show()
```

Insights:If one gender has significantly higher counts it might indicate a greater need for mental health support for that group.
Disparities might also suggest cultural or organizational differences that impact how genders report or experience mental health issues.

```python
crosstab = pd.crosstab(df['seek_help'], df['leave'])
fig = px.imshow(
    crosstab,
    title='Correlation Between Seeking Help and Accessibility to Medical Leave for Mental Health',
    color_continuous_scale='Blues',
    text_auto=True  )
fig.update_layout(
    xaxis_title='Medical Leave',
    yaxis_title='Seek Help',
    title_x=0.5,
    title_font=dict(size=30, family='Arial', color='red'),
    xaxis=dict(title=dict(font=dict(size=20, family='Arial', color='black'))),
    yaxis=dict(title=dict(font=dict(size=20, family='Arial', color='black')))
)
fig.show()
```

Insights:This information can be crucial for understanding whether employees who seek help tend to have more flexible medical leave policies, indicating an employer's supportive approach to mental health.
